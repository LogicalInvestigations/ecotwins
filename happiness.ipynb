## Perception

* Internal signals 
    * Interoception (Measured in SI units). Examples: glucose concentration, water concentration, body temp, pulse.
    * Proprioception (Measured as Booleans). Examples: leg_extended, arm_flexed, last_action_forward
* External signals (from the 5+ senses)
    * Raw perception (Measured in SI units). Examples: air temp, ground humidity, water salinity, brightness, colors, sound waves.
    * Object perception (Measured as a list of relative positions for each object type).
Example: Wolves at [(2,4), (5,8)].
    
Some of these signals can be input to the happiness function.

```python
def utility(x):
    return 1-np.exp(-0.5*x)
              
x = np.linspace(0,10)
y = utility(x)
plt.plot(x, y)
plt.xlabel("Energy")
plt.ylabel("Energy utility")
plt.yticks([0,1])
plt.title('y=1-np.exp(-0.5*x)')
;
```

## Happiness in general

Example of a happiness function with 3 variables: 

$$Happiness = Glucose \cdot Water \cdot (0.6 + 0.4 \cdot Oxytocin).$$

More generally we can do this:

$$Happiness= \Pi_x (a_x + (1-a_x)x),$$

where $0 \leq a_x \leq 1$ and $x$ is a happiness variable. This gives one parameter $a_x$ per variable $x$. 

We may want to use evolutionary algorithms to find good coefficients $a_x$.

More generally, we could model happiness as a (quadratic) polynomial in the perception variables.

Thus we can handle attraction/repulsion curves, bell curves, and food smell when hungry.



## Preprocessing interoception and terrain signals

1. Normalize: subtract the raw value by the mean and divide by sd.
1. Standardize: apply sigmoid.

```python
mu = 37
sd = 1
def normalize(x):
    return (x-mu)/sd
```

```python
normalize(39)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
              
x = np.linspace(-10,10)
y = sigmoid(x)
plt.plot(x, y)
plt.xlabel("Normalized signal")
plt.ylabel("Perception")
plt.yticks([0,1])
plt.title('$y=1/(1+e^{-x})$')
;
```

```python
def stimulus(x):
    return sigmoid(normalize(x))

stimulus(38)
```
